```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load the dataset
df = pd.read_excel('Road.xlsx')
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values by forward-filling
df.ffill(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))
    label_encoders[column] = le
```

```python
# Select relevant features based on domain knowledge
features = ['Age_band_of_driver', 'Sex_of_driver', 'Educational_level', 'Driving_experience', 
            'Type_of_vehicle', 'Road_surface_type', 'Road_surface_conditions', 'Light_conditions', 
            'Weather_conditions', 'Type_of_collision', 'Number_of_vehicles_involved', 
            'Number_of_casualties', 'Vehicle_movement', 'Casualty_class', 'Sex_of_casualty', 
            'Age_band_of_casualty', 'Casualty_severity']

X = df[features]
y = df['Accident_severity']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = knn.predict(X_test)

# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(knn, X_test, y_test, n_repeats=10, random_state=42)
importance = result.importances_mean

for i, imp in enumerate(importance):
    print(f"Feature {features[i]}: {imp}")
```

```python
import joblib

# Save the model
joblib.dump(knn, 'knn_road_accident_model.pkl')

# Load the model
# loaded_model = joblib.load('knn_road_accident_model.pkl')
```

```python
# Save the predictions to a CSV file
df_results = pd.DataFrame({
    'Actual_Severity': y_test,
    'Predicted_Severity': y_pred
})

df_results.to_csv('predictions.csv', index=False)
print("Predictions saved successfully!")
```

```python
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importance
})
feature_importance_df.to_csv('feature_importance.csv', index=False)
```
